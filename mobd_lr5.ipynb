```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
```

```python
spark = SparkSession \
        .builder \
        .appName("trvnq-mobd-lab5") \
        .getOrCreate()

print(spark.sparkContext.uiWebUrl)
```

```python
df = spark.read.csv('/content/csgo_players.csv', header=True)
df.show(5)
```

```python
df.printSchema()
```

```python
df.groupby('country') \
    .agg(F.count('*').alias('player_cnt')) \
    .show(5)
```

```python
df.select('*') \
    .where(F.col('total_kills') > 17000) \
    .where(F.col('total_deaths') < 13000) \
    .show(5)
```

```python
df.filter((F.col('country') == 'Russia') | (F.col('country') == 'France')) \
         .filter(F.col('rifle_kills') > 5000) \
         .select('nickname', 'country', 'rifle_kills', 'total_kills') \
         .show()
```
