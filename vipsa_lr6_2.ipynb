```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
cylinder_bands = fetch_ucirepo(id=32)

# data (as pandas dataframes)
X = cylinder_bands.data.features
y = cylinder_bands.data.targets

# metadata
print(cylinder_bands.metadata)

# variable information
print(cylinder_bands.variables)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame(X['ink temperature'])
df
```

```python
#-	выполнить визуализацию данных ink temperature
```

```python

plt.figure()
sns.histplot(df)
plt.grid()
plt.show()
```

```python


y = df
t =  np.arange(len(y))
plt.figure(figsize=(12,5))
plt.plot(t,y)
plt.grid()
plt.show()
```

```python
window = 50
rolling_mean = pd.Series(y['ink temperature']).rolling(window).mean()

plt.figure(figsize=(12,5))
plt.plot(t, y, alpha=0.5)
plt.plot(t, rolling_mean, linewidth=2)
plt.legend()
plt.grid()
plt.show()
```

```python
from statsmodels.tsa.stattools import acf

acf_y = acf(y['ink temperature'].dropna().values)

plt.figure(figsize=(10,4))
plt.stem(acf_y)
plt.grid()
plt.show()

print(acf_y[:30])


```
