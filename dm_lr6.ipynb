```python
def recursive(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return -3 * recursive(n-1) + 4 * recursive(n-2) + 3 * (2 ** n) + (n + 3)

def formul4(n):
    return -2.392 + 0.392 * ((-4) ** n) + 2**(n+1) + 0.1 * (n ** 2) + 0.86 * n


print(recursive(5))
print(formul4(5))
print(recursive(11))
print(formul4(11))
print(recursive(17))
print(formul4(17))
```
