```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('vipsa-lr1-db.csv', sep=';')
x = df['v021'].str.replace(',', '.').astype(float)
y = df['v022'].str.replace(',', '.').astype(float)
z = df['v023'].str.replace(',', '.').astype(float)
d = df['v024'].str.replace(',', '.').astype(float)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(x, edgecolor='black')

plt.figure(figsize=(10, 5))
plt.boxplot(x)

plt.figure(figsize=(10, 5))
plt.scatter(x.index, x.values, s=10)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(y, edgecolor='black')

plt.figure(figsize=(10, 5))
plt.boxplot(y)

plt.figure(figsize=(10, 5))
plt.scatter(y.index, y.values, s=10)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(z, bins=20, edgecolor='black')

plt.figure(figsize=(10, 5))
plt.boxplot(z)

plt.figure(figsize=(10, 5))
plt.scatter(z.index, z.values, s=10)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(d, bins=20, edgecolor='black')

plt.figure(figsize=(10, 5))
plt.boxplot(d)

plt.figure(figsize=(10, 5))
plt.scatter(d.index, d.values, s=10)
```

```python
print(f" : {x.shape}")
print(f": {x.var()}")
print(f": {x.mean()}")
print(f": {x.median()}")
print(f" : {x.quantile(0.75)}")
print(f" : {x.quantile(0.25)}")
print(f" : {x.skew()}")
print(f" : {x.kurt()}")
print(f" : {x.min()}")
print(f" : {x.max()}")
print(f" : {x.max() - x.min()}")
print(f": {x.mode()}")
```
